<img align="left" src="./image/Classification.png"     style=" width:380px; padding: 10px; " > Sınıflandırma problemlerine örnek olarak, bir e-postanın Spam veya Spam Değil olarak tanımlanması ya da bir tümörün kötü huylu (malignant) veya iyi huylu (benign) olup olmadığının belirlenmesi verilebilir. Özellikle, bunlar iki olası sonucun bulunduğu *ikili (binary) sınıflandırma* örnekleridir.

Sonuçlar, 'pozitif'/'negatif' çiftleri şeklinde tanımlanabilir, örneğin: 'evet'/'hayır', 'doğru'/'yanlış' veya '1'/'0'.

Sınıflandırma veri kümelerine ait grafiklerde genellikle semboller kullanılarak örneklerin sonuçları gösterilir. Aşağıdaki grafiklerde, `'X'` pozitif değerleri, `'O'` ise negatif sonuçları temsil etmektedir.


```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
pos = y_train == 1
neg = y_train == 0

fig,ax = plt.subplots(1,2,figsize=(8,3))
#plot 1, single variable
ax[0].scatter(x_train[pos], y_train[pos], marker='x', s=80, c = 'red', label="y=1")
ax[0].scatter(x_train[neg], y_train[neg], marker='o', s=100, label="y=0", facecolors='none',
              edgecolors=dlc["dlblue"],lw=3)

ax[0].set_ylim(-0.08,1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('one variable plot')
ax[0].legend()

#plot 2, two variables
plot_data(X_train2, y_train2, ax[1])
ax[1].axis([0, 4, 0, 4])
ax[1].set_ylabel('$x_1$', fontsize=12)
ax[1].set_xlabel('$x_0$', fontsize=12)
ax[1].set_title('two variable plot')
ax[1].legend()
plt.tight_layout()
plt.show()
```

 Grafiklerde dikkat edilmesi gereken noktalar:

- Tek değişkenli grafikte, pozitif sonuçlar kırmızı `'X'` sembolleriyle gösterilir ve `y=1` olarak bulunur. Negatif sonuçlar ise mavi `'O'` sembolleriyle gösterilir ve `y=0` konumundadır.
   - Hatırlayın, doğrusal regresyon (linear regression) durumunda `y` yalnızca iki değerle sınırlı olmazdı, herhangi bir değer alabilirdi.

- İki değişkenli grafikte `y` ekseni mevcut değildir. Pozitif sonuçlar kırmızı `'X'` sembolleriyle gösterilirken, negatif sonuçlar mavi `'O'` sembolleriyle temsil edilir.
   - Hatırlayın, çok değişkenli doğrusal regresyon durumunda `y` yalnızca iki değerle sınırlı olmazdı ve benzer bir grafik üç boyutlu olabilirdi.


## Doğrusal Regresyon Yaklaşımı

Geçtiğimiz hafta, doğrusal regresyon kullanarak bir tahmin modeli oluşturmuştunuz. Şimdi, derste anlatılan basit bir örnek üzerinden bu yaklaşımı burada deneyelim. Model, tümör boyutuna bağlı olarak bir tümörün iyi huylu (benign) ya da kötü huylu (malignant) olup olmadığını tahmin edecektir.

Şu adımları deneyin:

- 'Doğrusal Regresyonu Çalıştır' düğmesine tıklayarak verilen veri için en iyi doğrusal regresyon modelini bulun.
    - Ortaya çıkan doğrusal modelin **verilere iyi uymadığını** unutmayın.

Sonuçları iyileştirmenin bir yolu *eşik (threshold)* uygulamaktır.

- '0.5 Eşiğini Aç/Kapat' kutusunu işaretleyerek eşik uygulandığında tahminlerin nasıl göründüğünü inceleyin.
    - Bu tahminler iyi görünmektedir, çünkü tahminler verilerle eşleşmektedir.

- *Önemli:* Şimdi, büyük tümör boyutu aralığında (yaklaşık `10` civarında) ek 'kötü huylu (malignant)' veri noktaları ekleyin ve doğrusal regresyonu tekrar çalıştırın.
    - Model artık büyük tümörleri doğru tahmin edebiliyor, ancak `x=3` noktasındaki veri yanlış tahmin ediliyor!

- Grafiği temizlemek veya yenilemek için, grafiği oluşturan hücreyi tekrar çalıştırın.
